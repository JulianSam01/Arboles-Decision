## Establecer las variables almacenadas que contengan los datos que entrenan y testearan el modelo 

```python
base_predicciones = ['volatile acidity','citric acid','total sulfur dioxide','pH','sulphates']
variable_a_predecir = ['quality']
```

```python
predicciones = Datos[base_predicciones]
predecir = Datos[variable_a_predecir]
```

```python
predicciones
```

```python
predecir
```

## Miramos la correlacion entre variables 

```python
correlacion = predicciones.corr()
correlacion
```

```python
sns.heatmap(correlacion)
plt.title('Matriz de Correlación')
plt.show()
```

## - volatile acidity y citric acid: Tienen una correlación negativa fuerte de -0.552, lo que sugiere que, en general, a medida que aumenta la acidez volátil, disminuye el ácido cítrico. Este comportamiento puede influir en la percepción de frescura del vino.
## - citric acid y pH: La correlación negativa de -0.542 indica que a mayor acidez (más bajo el valor de pH), mayor es la presencia de ácido cítrico. Esto es importante, ya que el pH y la acidez son factores de calidad y frescura en el vino.
## - citric acid y sulphates: Tienen una correlación positiva moderada de 0.313, lo cual indica que los vinos con más ácido cítrico tienden a tener también más sulfitos. Esto puede contribuir a la estabilidad y preservación del vino.
## - volatile acidity y pH: Correlación positiva moderada de 0.235, sugiriendo que la acidez volátil podría aumentar cuando el pH es más alto, lo que puede impactar negativamente en la calidad percibida del vino.

## Creamos el modelo y seleccionar la cantidad de datos usados en entrenar y testear el modelo 

```python
x_train, x_test, y_train, y_test = train_test_split(predicciones, predecir, test_size = 0.15, random_state = 13)
```

```python
tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit(x_train, y_train)
```

```python
plot_tree(arbol)
```

```python
predicciones=arbol.predict(x_test)
```

```python
predicciones
```

## Examinamos la cantidad de datos en una matriz de confusion 

# Esta matris se encarga de comparar los datos reales con los predichos por el modelo

```python
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames = ['Actual'], colnames = ['Predicciones'])
```

## Ahora evaluamos el rendimiento del modelo 

```python
accuracy = accuracy_score(y_test, predicciones)
accuracy
```

```python
precision = precision_score(y_test, predicciones, average = 'weighted')
precision
```

```python
recall = recall_score(y_test, predicciones, average = 'weighted')
recall
```

```python
f1 = f1_score(y_test, predicciones, average = 'weighted')
f1
```
